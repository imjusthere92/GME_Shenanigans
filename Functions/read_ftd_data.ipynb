```python
import os
import requests
import pandas as pd
import numpy as np
import json

from pathlib import Path
import csv

from dotenv import load_dotenv
load_dotenv()

import pandas_datareader as web
import datetime as dt

import panel as pn
from panel.interact import interact
import plotly.express as px
pn.extension("plotly")

%matplotlib inline
```

```python
def import_df(csv_path):
    df = pd.read_csv(
    Path(csv_path),
    infer_datetime_format=True,
    parse_dates=True,
    index_col="Date",
    )
    return df 
```

```python
gme_csv_path = '../FilesExport/GME_all_data.csv'
gme_df = import_df(gme_csv_path)
gme_df
```

```python
gme = gme_df
```

```python
gme_new = gme['high'] - gme['low']
gme_new

gme['price_change'] = gme_new
```

```python
change_percent_of_price = gme['price_change'] / gme['close'] 
change_percent_of_price = change_percent_of_price * 100
gme['change_percent_of_price'] = change_percent_of_price
```

```python
gme
```

```python
sort_percent_change = gme.sort_values('change_percent_of_price',ascending=False)
```

```python
sort_percent_change = sort_percent_change[['GME_QUANTITY_FAILS','change_percent_of_price','close']]
```

```python
sort_ftd_high_to_low = gme.sort_values('GME_QUANTITY_FAILS',ascending=False)
sort_ftd_high_to_low = sort_ftd_high_to_low[['GME_QUANTITY_FAILS','change_percent_of_price']]
sort_ftd_high_to_low.rename(columns={'GME_QUANTITY_FAILS':'GME_FTD'},inplace=True)
```

```python
sort_percent_change.reset_index(inplace=True)
sort_percent_change.rename(columns={'Date':'Date_Price_Change'},inplace=True)

sort_ftd_high_to_low.reset_index(inplace=True)
sort_ftd_high_to_low.rename(columns={'Date':'Date_FTD'},inplace=True)
```

```python
merged_df = sort_percent_change.merge(sort_ftd_high_to_low, how='inner',right_index=True, left_index=True)
merged_df
```
